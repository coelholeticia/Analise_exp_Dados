```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
round(df["Custo"].sum(), 2)
```

```python
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
round(df["Lucro"].sum(), 2)
```

```python
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"] 
```

```python
df.head()
```

```python
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
df["Tempo_envio"].dtype
```

```python
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
df.isnull().sum()
```

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format #formatação de valores 
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de Produtos vendidos", color="blueviolet")
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().plot.bar(title="Lucro x Ano", color="plum")
plt.xlabel('Ano')
plt.xlabel('Receita');
```

```python
df.to_csv("ANalise_vendas", index=False)
```
